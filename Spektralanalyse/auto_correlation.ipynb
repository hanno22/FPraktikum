```python
import numpy as np
import matplotlib.pyplot as plt

# Generate pure sinus signal
t = np.linspace(0, 6, 1000)
signal = np.sin(2 * np.pi * t)
noise = np.random.normal(0, 1, 1000)
```

```python
def plot_autocorr(ts, ys):
    fig, (ax1, ax2) = plt.subplots(2,1)
    ax1.plot(ts, ys)
    ax1.set(xlabel='t/s', ylabel='A')
    ax1.grid()
    corr= np.correlate(ys, ys, mode='same')
    ax2.plot(ts - np.mean(ts), corr, 'r--', label='Autocorrelation')
    ax2.grid()
    ax2.set(xlabel='$\\tau / \mathrm{s}$', ylabel='A')
    fig.tight_layout()
    return fig
```

```python
fig1 = plot_autocorr(t, signal)
fig1.savefig('out/pure_sine_autocorr')
fig2 = plot_autocorr(t, noise)
fig2.savefig('out/pure_noise_autocorr')

corr_signal = np.zeros_like(t)
corr_length = int(1000 / 6)
for i in range(len(corr_signal)):
    if(i < corr_length):
        corr_signal[i] = np.random.normal()
    else:
        corr_signal[i] = corr_signal[i - corr_length] + i * 0.001# np.random.normal()

fig3 = plot_autocorr(t, corr_signal)
```
