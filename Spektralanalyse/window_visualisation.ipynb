```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.signal as sgn
from scipy.fft import fft, fftfreq
```

```python
dt = 1 / 500
N = 3000
ts = np.arange(0, dt * N, dt)
```

```python
signal = (2*np.sin( 2 * np.pi * ts * 15) + 3*np.sin( 2 * np.pi * ts * 14))
signal += 10* np.random.normal(size=N)

signal *= sgn.windows.hamming(N)

yfft = fft(signal)
yfft =  2.0 / N * np.abs(yfft[0:N//2])
yfft[0] = yfft[0] / 2
freqs = fftfreq(N, d=dt)[0:N//2]
```

```python
fig, ax = plt.subplots(2, 1, figsize = [10,10])
ax[0].plot(ts, signal)
ax[0].set(xlabel="t / s")
ax[0].grid()

ax[1].plot(freqs, yfft)
ax[1].set(xlabel="f / Hz", xlim = [0,20])
ax[1].grid()
```
