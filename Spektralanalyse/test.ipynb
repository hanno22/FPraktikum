```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
N = int(1e6)
data = np.random.normal(size=N)

num_bits = 4
quantization_levels = 2 ** num_bits
quantization_step = 2 / quantization_levels
quantized_data = np.around(data / quantization_step) * quantization_step
```

```python
def gaussian_function(x, mean, std_dev):
    exponent = -(x - mean) ** 2 / (2 * std_dev ** 2)
    return np.exp(exponent) / (std_dev * np.sqrt(2 * np.pi))
```

```python
fig, (ax, ax1) = plt.subplots(2, 1, sharex=True)#, sharey=True, sharex=True)
ax.hist(data, 50, density=True)#, rwidth=0.8)
ax1.hist(quantized_data, 51, density=True)#, rwidth=0.8)

xs = np.linspace(min(data), max(data), 1000)
ax.plot(xs, gaussian_function(xs,0, 1), 'r--')
ax1.plot(xs, gaussian_function(xs,0, 1), 'r--')
ax1.set(xlabel="Z", ylabel="P(Z)", title='quantized data')
ax.set(ylabel="P(Z)", title='continues data')

fig.savefig('out/quant_noise')
```
