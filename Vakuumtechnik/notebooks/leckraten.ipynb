```python
import numpy as np
import matplotlib.pyplot as plt
scale = 133.322
```

```python
data = np.loadtxt("../data/Leckraten.txt", skiprows=1)
t = data[:,0]
rate1 = data[:,1] * scale
rate2 = data[:,2] * scale
```

```python
fig, ax = plt.subplots()

def plot(ydata, label):
       fit_params = np.polyfit(t, ydata, 1)
       fit_function = np.poly1d(fit_params)
       ts = np.linspace(np.min(t), np.max(t), 1000)
       ax.scatter(t, ydata, label=label)
       ax.plot(ts, fit_function(ts), label =f"$P(t) = t \\cdot {fit_params[0]:.2} \\mathrm{{Pa}}/\\mathrm{{s}} + {fit_params[1]:.2}\\mathrm{{Pa}}$")
       print( f'Leckrate {6.5 * fit_function[0]}')


plot(rate1, "Vor Versuchen")
plot(rate2, "Nach Versuchen")

ax.set(xlabel="$t/\\mathrm{s}$",
       ylabel="$P / \\mathrm{Pa}$")
ax.grid()
ax.legend()
```

```python
fig.savefig('../out/leckraten.png')
```
