```python
import numpy
import numpy as np
import matplotlib.pyplot as plt
```

```python
data = np.loadtxt('data/Kennlinie.csv', skiprows=1)
```

```python
fig, ax = plt.subplots()
voltage = data[:, 0]
current = data[:, 2]

ax.plot(voltage[9:], current[9:], marker='x', label='steigend')
ax.plot(voltage[0:9], current[0:9], marker='x', label='fallend')
ax.grid()
ax.set(xlabel="$U_\mathrm{Total} / \\mathrm{V}$", ylabel="$I / \\mathrm{mA}$")
ax.legend()

fig.savefig('out/Kennlinie_Thyratron')
```
