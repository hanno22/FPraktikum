```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
data = np.loadtxt('data/210417_Halbleiter/kalibration_mittagspause.TKA', skiprows=2)
channels = np.arange(2, len(data) + 2)
E_max = 5.486
def gauss_function(x, a, x0, sigma):
    return a * np.exp(-(x - x0)**2 / (2 * sigma**2))
```

```python
fig, ax = plt.subplots()
channel_max = channels[np.argmax(data)]
print(channel_max)
def channel2Energie(c):
    return c * E_max / channel_max;

ax.scatter(channel2Energie(channels), data, 0.5, marker='.')
ax.set(xlim=[5,5.6], ylabel="I/AU", xlabel="E / MeV")
old_ticks = list(ax.get_xticks())
del old_ticks[np.argmin(np.abs( old_ticks - channel2Energie(channel_max)))]
ax.set_xticks( old_ticks + [(channel2Energie(channel_max))])
ax.set_yticklabels([])
ax.grid()



print(E_max / channel_max)


```
