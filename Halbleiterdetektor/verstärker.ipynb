```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
data = np.loadtxt('data/210417_Halbleiter/VV_Impulsform.csv', delimiter=',', usecols=[3,4,9,10])
fig, ax = plt.subplots(2,1, sharex=True)
ax[0].plot(data[:,0] * 1e3, data[:,1] / 10 *1e3, label = "ohne Verstärker")
ax[1].plot(data[:,2] * 1e3, data[:,3] / 10 *1e3, label = "Vorverstärker")
ax[1].set(xlabel='$t / ms$', ylabel='$U / mV$')
ax[0].set(ylabel='$U / mV$')
ax[0].grid()
ax[1].grid()
ax[0].set_yticks([0, data[:,1][np.argmax(np.abs( data[:,1]))]/10 *1e3])
ax[1].set_yticks([0, data[:,3][np.argmax(np.abs( data[:,3]))]/10*1e3])
ax[0].legend()
ax[1].legend()
fig.tight_layout()
fig.savefig('out/Vorverstärker.png')
```

```python
data = np.loadtxt('data/210417_Halbleiter/HV_Impulsform.csv', delimiter=',', usecols=[3,4,9,10])
fig, ax = plt.subplots(2,1, sharex=True)
ax[1].plot(data[:,0] * 1e3, data[:,1] / 10, label = "Hauptverstärker")
ax[0].plot(data[:,2] * 1e3, data[:,3] / 10 *1e3, label = "Vorverstärker")
ax[1].set(xlabel='$t / ms$', ylabel='$U / V$')
ax[0].set(ylabel='$U / mV$')
ax[0].grid()
ax[1].grid()
ax[1].set_yticks([0, data[:,1][np.argmax(np.abs( data[:,1]))]/10])
ax[0].set_yticks([0, data[:,3][np.argmax(np.abs( data[:,3]))]/10*1e3])
ax[0].legend()
ax[1].legend()
fig.tight_layout()
fig.savefig('out/Hauptverstärker.png')
```
