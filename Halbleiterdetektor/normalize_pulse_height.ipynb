```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
data_ph = np.loadtxt('data/pulse_height.csv', skiprows=1, delimiter=',')
data_normalize = np.loadtxt('data/pulse_height.csv', skiprows=1, delimiter=',')
```

```python
fig, (ax, ax2) = plt.subplots(2, 1)
ax.scatter(data_ph[:,0], data_ph[:,1])
ax.set(xlabel="PulseHeight", ylabel= "$U/\mathrm{mV}$")
ax.grid()
ax2.scatter(data_normalize[:,0], data_normalize[:,1])
ax2.set(xlabel="Normalize", ylabel= "$U/\mathrm{mV}$")
ax2.grid()
fig.tight_layout()
```
