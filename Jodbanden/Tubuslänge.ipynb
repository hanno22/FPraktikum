```python
import numpy as np
from os import walk
import matplotlib.pyplot as plt
```

```python
path = 'data/Tubusl√§nge/'
filenames = next(walk(path))[2]
data = {}
for filename in filenames:
    data[filename] = np.loadtxt(path + filename)
```

```python
labels = {
    "1.dat": 0,
    "4.dat": 0.5,
    #"2.dat": 1,
    "6.dat": 1.25,
    "5.dat": 1.5,
    "3.dat": 2
}


fig, ax = plt.subplots()
for d in labels:
    ax.plot(data[d][:,0], data[d][:,1], label = f"{labels[d]}")
ax.grid()
ax.legend()
ax.set(xlabel="Steps", ylabel="I/AU")
fig.savefig("out/Tubuslaenge.svg")
```
