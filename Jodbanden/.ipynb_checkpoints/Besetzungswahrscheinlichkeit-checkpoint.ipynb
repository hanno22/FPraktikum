```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.constants as c
```

```python
def E(n):
    D = 1.54 * c.e
    omega_0 = 1.32e4 * c.c
    return c.h * omega_0 *(n + 1/2) - (c.h * omega_0) ** 2 /(4*D) * (n + 1/2)**2

T = 300

ns = np.arange(0, 189, 1)
Es = E(ns)

Z = np.sum(np.exp(-Es / (c.k * T)))
ps = 1 / Z * np.exp(-Es / (c.k * T))
```

```python
fig, ax =plt.subplots()
lns1 = ax.bar(ns, ps, label="$p(n)$")
ax2 = ax.twinx()
lns2 = ax2.bar(ns, Es / c.e, color='tab:red', label="$E(n)$", alpha=0.5)
ax2.set(ylabel="$E / \\mathrm{eV}$", ylim=[0,0.2], yticks=Es[0:10]/c.e)
ax2.grid(color='tab:red', linestyle="--")
ax.set(xlim=[-0.5,9.5], xticks=np.arange(0,10,1), yticks=ps[0:6], xlabel="$n$", ylabel="$p(n)$")
ax.grid(color='tab:blue', axis="y", linestyle="--")

print(lns1)
print(lns2)
lns = [lns1]+[lns2]
labs = [l.get_label() for l in lns]
ax.legend(lns, labs, loc=0)
fig.savefig("out/besetzungswahrscheinlichkeit.svg")
```
