```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
resonanz = lambda omega, omega_0, gamma, CR : 1.0 / (np.sqrt((omega_0 ** 2 - omega ** 2) ** 2 + gamma ** 2 * omega ** 2)) #* omega * CR / np.sqrt(1.0 + (omega * CR) ** 2)

omegas = np.linspace(0,3, 1000)
fig, ax = plt.subplots()
omega_max = []
amp_max = []
for gamma in np.arange(0.3, 1.1, 0.1):
    ax.plot(omegas, resonanz(omegas, 1.0, gamma, 1), label = f"$\\gamma/\\omega_0 = {gamma:.2}$")

for gamma in np.linspace(0.25, 5.0, 1000):
    res = resonanz(omegas, 1.0, gamma, 1)
    omega_max.append(omegas[np.argmax(res)])
    amp_max.append(np.max(res))
ax.plot(omega_max, amp_max, "k--", label="Max")
ax.set(xlabel="$\\omega/\\omega_0$", ylabel="$R(\\omega)$")
ax.grid()
ax.legend()
```
