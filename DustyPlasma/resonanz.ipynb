```python
import numpy as np
import matplotlib.pyplot as plt
import scipy as scp
```

```python
data = np.loadtxt("data/Resonanzmethode/resonanz2/.dy")
k = 9
x = data[:,0]
y = data[:,1] * k
```

```python
#Fit:

fit_function = lambda omega, A, omega_0, gamma, CR : A / (np.sqrt((omega_0 ** 2 - omega ** 2) ** 2 + gamma ** 2 * omega ** 2)) * omega * CR / np.sqrt(1.0 + (omega * CR) ** 2)
p0 = (5.0, 11.0, 25.0, 5)
(A, omega_0, gamma, CR), cv = scp.optimize.curve_fit(fit_function, x, y, p0)
```

```python
fig, ax = plt.subplots()
ax.scatter(x, y)
xs = np.linspace(np.min(x), np.max(x), 1000)
ax.plot(xs, fit_function(xs, A, omega_0, gamma, CR))
ax.grid()
ax.set(xlabel="$f / \\mathrm{Hz}$", ylabel="$A / \\mathrm{\\mu m}$")

print(f'Omega_0={omega_0:.4}Hz, Gamma = {gamma : .4}Hz')
```
